```python
import igraph as ig
import pandas as pd
import numpy as np
import csv
import json
```

```python
# Define the file path for the JSON file
json_file_path = "deezer_europe\deezer_europe_features.json"

# Load the graph data from the JSON file
with open(json_file_path, "r") as f:
    data = json.load(f)

```

```python
#get the tab of edges
edge_path = "deezer_europe\deezer_europe_edges.csv"
with open(edge_path, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    next(reader)
    row_count = sum(1 for row in reader)
    csvfile.seek(0)  # reset the file pointer
    col_count = len(next(reader))
    edges = np.zeros([row_count, col_count])
    for i, row in enumerate(reader):
        print(', '.join(row))
        edges[i,:]= row
```

```python
#get the tab of targets
target_path = "deezer_europe\deezer_europe_target.csv"
with open(target_path, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    next(reader)
    row_count = sum(1 for row in reader)
    csvfile.seek(0)  # reset the file pointer
    col_count = len(next(reader))
    target = np.zeros([row_count, col_count])
    for i, row in enumerate(reader):
        print(', '.join(row))
        target[i,:]= row
```

```python
# Create an empty graph
g = ig.Graph()

# Add vertices
for v in data["vertices"]:
    g.add_vertex(name=v["name"], **v["attributes"])

# Add edges
for e in edges:
    g.add_edge(e[0], e[1])
```
